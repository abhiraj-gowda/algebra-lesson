### graph for cross product
let the points be [6,10] and [4,5]

```python
# cross product 
vv1=np.array([6,10])
vv2=np.array([4,5])
cross_pro=np.cross(vv1,vv2)
dot_pro=np.dot(vv1,vv2)
print(cross_pro,'\n',dot_pro)
```

```python
# projection of 'a' vector over 'v' vector
a=np.array([6,10])
v=np.array([12,0])
magnitudev= np.sqrt(sum(v**2))
proj_a_v = (np.dot(a, v) / magnitudev**2)*v
print(proj_a_v)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create some sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Set the style to whitegrid
sns.set(style="whitegrid",palette="muted")

# Create a figure with four subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Set the context to paper
sns.set_context("paper")
axs[0, 0].plot(x, y)
axs[0, 0].set_title("Paper Context")

# Set the context to notebook
sns.set_context("notebook")
axs[0, 1].plot(x, y)
axs[0, 1].set_title("Notebook Context")

# Set the context to talk
sns.set_context("talk")
axs[1, 0].plot(x, y)
axs[1, 0].set_title("Talk Context")

# Set the context to poster
sns.set_context("poster")
axs[1, 1].plot(x, y)
axs[1, 1].set_title("Poster Context")

# Layout so plots do not overlap
fig.tight_layout()

plt.show()
```
